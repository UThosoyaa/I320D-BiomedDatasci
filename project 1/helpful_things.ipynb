## My Markdowns Won't Go Into the Next Line
---

For some reason (I don't really know why this is) markdown doesn't automatically go into a new line just because you pressed enter. See example here:  

Hi! This is one line.
Hello! This is the line after that.

However, it **will** do it for lists.

1) Hi! This is one line.
2) Hello! This is the line after that.

- Hi, this is one line.
- Hi, this is the line after that

You can do things like this where you put in two different new lines

Like so

But if you don't want to do that, you can just add two spacebars to the end of your sentence (  )  
and it'll make it into a newline.
